```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
cancer = load_breast_cancer()
```

```python
X, y = cancer.data, cancer.target
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
dtree = DecisionTreeClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
dtree.fit(X_train, y_train)
```

```python
dtree.score(X_train, y_train)
```

```python
dtree.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'max_depth': [3, 4, 5],
         'min_samples_split': [2, 3, 4, 5, 6]}
```

```python
grid = GridSearchCV(dtree, param_grid=params)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.score(X_train, y_train)
```

```python
grid.score(X_test, y_test)
```

```python
grid.best_params_
```

### Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
```

```python
forest.fit(X_train, y_train)
```

```python
forest.score(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
params = {'n_estimators': [20, 50, 100, 1000],
         'max_depth': [2, 3]}
```

```python
forest_grid = GridSearchCV(forest, param_grid=params, cv = 5)
```

```python
forest_grid.fit(X_train, y_train)
```

```python
forest_grid.score(X_train, y_train)
```

```python
forest_grid.score(X_test, y_test)
```

```python
forest_grid.best_params_
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
plot_tree(dtree, ax = ax);
```

```python
pd.DataFrame({'features':cancer.feature_names, 
              'importances': forest_grid.best_estimator_.feature_importances_})
```

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
```

```python
adaboost = AdaBoostClassifier()
```

```python
adaboost.fit(X_train, y_train)
```

```python
adaboost.score(X_train, y_train)
```

```python
adaboost.score(X_test, y_test)
```

```python
ada_grid = GridSearchCV(adaboost, param_grid={'n_estimators': [50, 100, 200]})
```

```python
ada_grid.fit(X_train, y_train)
```

```python
ada_grid.score(X_train, y_train)
```

```python
ada_grid.score(X_test, y_test)
```

```python
grad_boost = GradientBoostingClassifier()
```

```python
grad_boost.fit(X_train, y_train)
```

```python
grad_boost.score(X_train, y_train)
```

```python
grad_boost.score(X_test, y_test)
```

```python
params = {'learning_rate': [0.1, 0.5, 0.01],
         'max_depth': [1, 2, 3],
         'n_estimators': [50, 100, 500]}
```

```python
grad_grid = GridSearchCV(grad_boost, param_grid=params)
```

```python
grad_grid.fit(X_train, y_train)
```

```python
grad_grid.score(X_train, y_train)
```

```python
grad_grid.score(X_test, y_test)
```
