```python
class MakeMarkov:
    def __init__(self, n_component = 3, n_iter = 1000):
        self.n_components = n_component
        self.n_iter = n_iter
        self.model = hmm.GaussianHMM(n_components = self.n_components, n_iter=self.n_iter)
        
    def __str__(self):
        return f'HMM model with {n_component}'
    
    def train(self, X):
        self.fit_model_ = self.model.fit(X)
        return self.fit_model_
    
    def score(self, X):
        return self.fit_model_.score(X)
```

```python
makemarkov = MakeMarkov()
```

```python
makemarkov.train(X)
```

```python
makemarkov.score(X)
```

```python
import numpy as np
import os
```

```python
hmm_models = []
for file in os.listdir('sounds/'):
    subfolder = os.path.join('sounds', file)
    if not os.path.isdir(subfolder):
        continue
    label = subfolder[subfolder.rfind('/') + 1:]
    print(subfolder)
    X = np.array([])
    y_words = []
    for filename in [x for x in os.listdir(subfolder) if x.endswith('.wav')][:-1]:
        filepath = os.path.join(subfolder, filename)
        sample_freq, audio = wavfile.read(filepath)
        mfcc_features = features.mfcc(audio)
#         if len(X) == 0:
#             X = mfcc_features
#         else:
#             X = np.append(X, mfcc_features, axis = 0)
        X = mfcc_features
        y_words.append(label)
        hmm_trainer = MakeMarkov()
        hmm_trainer.train(X)
        hmm_models.append((hmm_trainer, label))
        hmm_trainer = None
```

```python
insounds = ['sounds/apple/apple15.wav',
'sounds/kiwi/kiwi15.wav',
'sounds/lime/lime15.wav',
'sounds/banana/banana15.wav',
'sounds/pineapple/pineapple15.wav',
'sounds/orange/orange15.wav',
'sounds/peach/peach15.wav']
```

```python
insounds[0].split('/')[1]
```

```python
for file in insounds:
    samp, audio = wavfile.read(file)
    mfcc_features = features.mfcc(audio)
    max_score = -np.inf
    output_label = None
    for model in hmm_models:
        hmm_model, label = model
        score = hmm_model.score(mfcc_features)
        if score > max_score:
            max_score = score
            output_label = label
    print('True: ', file.split('/')[1])
    print('Predict: ', output_label)
```
